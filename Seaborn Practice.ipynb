```python
import seaborn as sns
```

```python
ds = sns.load_dataset("tips")
```

```python
ds.head()
```

# tip - dependent feature
#all others - independent features
#Heatmap - Corelation between two features
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix between two discrete dimensions or event types.It is very important feature selection

```python
ds.corr()
```

```python
sns.heatmap(ds.corr())
```

```python
sns.jointplot("tip","total_bill",data = ds,kind ="kde")
```

```python
#PairPlot is a scatterplot
```

```python
sns.pairplot(ds,hue="sex")
```

```python
#Barplot - says relation between a categorical feature and a feature
sns.barplot(x="sex",y="total_bill",data=ds)
```
