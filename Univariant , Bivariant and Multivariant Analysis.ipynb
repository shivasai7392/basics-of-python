```python
import seaborn as sns
```

```python
ds = sns.load_dataset("iris")
```

```python
ds
```

```python
sns.heatmap(ds.isnull())
```

```python
import pandas as pd
import numpy as np
```

```python
ds_species_setosa = ds.loc[ds["species"]=="setosa"]
```

```python
ds_species_virginica = ds.loc[ds["species"]=="virginica"]
```

```python
ds_species_versicolor = ds.loc[ds["species"]=="versicolor"]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(ds_species_setosa["sepal_length"],np.zeros_like(ds_species_setosa["sepal_length"]))
plt.plot(ds_species_virginica["sepal_length"],np.zeros_like(ds_species_virginica["sepal_length"]))
plt.plot(ds_species_versicolor["sepal_length"],np.zeros_like(ds_species_versicolor["sepal_length"]))
plt.show()
```

```python
sns.FacetGrid(ds,hue="species",height=5).map(plt.scatter,"petal_length","sepal_width").add_legend()
plt.show()
```

```python
sns.pairplot(ds,hue="species")
plt.show()
```
